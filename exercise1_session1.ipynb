```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rs = np.random.RandomState(seed=1)

rs.random()
```

```python
def sample_avg_ctrs(probability_of_click, number_of_sessions):
    ctrs = rs.binomial(1, probability_of_click, number_of_sessions)
    avg_ctrs = np.mean(ctrs)
    return avg_ctrs
```

```python
ctrs_100 = sample_avg_ctrs(0.1, 100)
ctrs_100
```

```python
ctrs_10000 = sample_avg_ctrs(0.1, 10000)
ctrs_10000
```

```python
number_of_sessions_list = range(10, 10_000, 100)

avg_ctrs = []

for number_of_sessions in number_of_sessions_list:
    avg_ctr = sample_avg_ctrs(0.1, number_of_sessions)
    avg_ctrs.append(avg_ctr)
```

```python
print(number_of_sessions_list)
print(avg_ctrs)
```

```python
plt.plot(number_of_sessions_list, avg_ctrs)
plt.axhline(0.1, c="r")
```

```python
def sample_avg_revenue(mu, sigma, size):
    # ctrs = rs.binomial(1, probability_of_click, number_of_sessions)
    rs = np.random.normal(mu, sigma, size)
    avg_revenue_per_user = np.mean(rs)
    return avg_revenue_per_user
```

```python
number_of_user_list = range(10, 10_000, 100)

avg_revenues = []

for number_of_user in number_of_user_list:
    avg_revenue = sample_avg_revenue(mu, sigma, number_of_user)
    avg_revenues.append(avg_revenue)
```
