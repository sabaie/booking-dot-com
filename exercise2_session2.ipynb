```python
def sample_avg_ctrs(probability_of_click, number_of_sessions, rs):
    ctrs = rs.binomial(1, probability_of_click, number_of_sessions)
    avg_ctrs = np.mean(ctrs)
    return avg_ctrs
```

```python
num_samples = 100
# num_sessions = 500
num_sessions_list = [10, 30, 100, 200, 500]

PROB = 0.4

for num_sessions in num_sessions_list:

    rs = np.random.RandomState(seed=1)

    ctrs = [sample_avg_ctrs(PROB, num_sessions, rs) for _ in range(num_samples)]

    plt.figure()
    plt.hist(ctrs, bins=np.linspace(0,1,41))
    plt.axvline(np.mean(ctrs), color="r")
    plt.title(f"numer of sessions: {num_sessions}");
```

```python
num_samples_list = [5, 10, 50, 100, 1000, 10000]
num_sessions = 100

PROB = 0.4

for num_samples in num_samples_list:

    rs = np.random.RandomState(seed=1)

    ctrs = [sample_avg_ctrs(PROB, num_sessions, rs) for _ in range(num_samples)]

    plt.figure()
    plt.hist(ctrs, bins=np.linspace(0,1,41))
    plt.axvline(np.mean(ctrs), color="r")
    plt.title(f"numer of samples: {num_samples}, std: {np.std(ctrs)}");
```
